```python
cellular_data=all_states_tech['Cellular data plan for a smartphone or other mobile device'].value_counts(normalize=True).reset_index()
dial_up=all_states_tech['Dial-up service'].value_counts(normalize=True).reset_index()
laptop=all_states_tech['Laptop or desktop'].value_counts(normalize=True).reset_index()
internet_service=all_states_tech['Access to the Internet'].value_counts(normalize=True).reset_index()
sat_internet_service= all_states_tech['Satellite Internet service'].value_counts(normalize=True).reset_index()
smartphone=all_states_tech['Smartphone'].value_counts(normalize=True).reset_index()
tablet=all_states_tech['Tablet or other portable wireless computer'].value_counts(normalize=True).reset_index()
```

```python
internet_service_yes = internet_service[internet_service['index']==1.0].values[0][1] +internet_service[internet_service['index']==2.0].values[0][1]
Yes = [tablet[tablet['index']==1.0].values[0][1],
dial_up[dial_up['index']==1.0].values[0][1],
laptop[laptop['index']==1.0].values[0][1],
internet_service_yes,
sat_internet_service[sat_internet_service['index']==1.0].values[0][1],
smartphone[smartphone['index']==1.0].values[0][1],
tablet[tablet['index']==1.0].values[0][1]]
No = [tablet[tablet['index']==2.0].values[0][1],
dial_up[dial_up['index']==2.0].values[0][1],
laptop[laptop['index']==2.0].values[0][1],
internet_service[internet_service['index']==3.0].values[0][1],
sat_internet_service[sat_internet_service['index']==2.0].values[0][1],
smartphone[smartphone['index']==2.0].values[0][1],
tablet[tablet['index']==2.0].values[0][1]]
Yes[:] = [x * 100 for x in Yes]
No[:] = [x * 100 for x in No]
```

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure

output_notebook()
fruits = ['Cellular', 'Dial-up', 'Laptop', 'Misc. Internet Services', 'Sat. Internet Services', 'SmartPhones','Tablets']
years = ['Yes','No']
colors = ["#006400", "#DC143C"]

data = {
    'fruits': fruits,
'Yes' : Yes,
'No' : No
        }

p = figure(x_range=fruits,plot_width=700 ,plot_height=500, title="Fruit Counts by Year",
           toolbar_location=None, tools="hover", tooltips="$name @fruits: @$name %")

p.vbar_stack(years, x='fruits', width=0.9, color=colors, source=data,
             legend_label=years)

p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.location = "top_left"
p.legend.orientation = "horizontal"
p.xaxis.major_label_orientation =  np.pi/4
show(p)
```

# Maps

```python


from bokeh.io import output_notebook, show, output_file
from bokeh.plotting import figure
from bokeh.models import GeoJSONDataSource, LogColorMapper, ColorBar
from bokeh.palettes import brewer
from bokeh.models import Slider, HoverTool

geosource = GeoJSONDataSource(geojson = json_data)

palette = brewer['Spectral'][8]

palette = palette[::-1]

color_mapper = LogColorMapper(palette = palette)
 
color_bar = ColorBar(color_mapper=color_mapper, label_standoff=8,width = 500, height = 20,
border_line_color=None,location = (0,0), orientation = 'horizontal')

TOOLS = "pan,wheel_zoom,reset,hover,save"


p =figure(title = 'USA Map with Mean Household Income, Mean Family Income, Population', tools=TOOLS,plot_height = 600 , plot_width = 950
)
p.xgrid.grid_line_color = None

p.ygrid.grid_line_color = None

p.patches('xs','ys', source = geosource,fill_color = {'field' :'population', 'transform' : color_mapper},
          line_color = 'black', line_width = 0.75, fill_alpha = 1,)
p.add_layout(color_bar, 'below')

hover = p.select_one(HoverTool)
hover.point_policy = "follow_mouse"
hover.tooltips=[
    ("Mean Family Income", "@family_income"),
    ("Mean Household Income", "@household_income"),
    ("Population", "@population"),
    ("State Name","@NAME")
]


output_notebook()
#Display figure.
show(p)


#Source - https://towardsdatascience.com/a-complete-guide-to-an-interactive-geographical-map-using-python-f4c5197e23e0
```

```python
from bokeh.io import output_file, show
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure
from bokeh.sampledata.commits import data
from bokeh.transform import jitter

output_file("categorical_scatter_jitter.html")

DAYS = ['Sun', 'Sat', 'Fri', 'Thu', 'Wed', 'Tue', 'Mon']

source = ColumnDataSource(data)

p = figure(plot_width=800, plot_height=300, y_range=DAYS, x_axis_type='datetime',
           title="Commits by Time of Day (US/Central) 2012-2016")

p.circle(x='time', y=jitter('day', width=0.6, range=p.y_range),  source=source, alpha=0.3)

p.xaxis.formatter.days = ['%Hh']
p.x_range.range_padding = 0
p.ygrid.grid_line_color = None

show(p)
```
